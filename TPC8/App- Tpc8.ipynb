```python
def distribui(bd,ind):
    dis=tpc.distrib(bd,ind)
    res=[]
    for v in dis:
        if v=='True':
            res.append("Apto"+": "+str(dis[v]))
        elif v=='False':
            res.append("Não Apto"+": "+str(dis[v]))
        res.append(str(v)+": "+str(dis[v]))
    return res

import TPC8 as tpc
import PySimpleGUI as sg
sg.theme('DarkTeal11')
BD=tpc.lerDataset("emd.csv")
res=[]
interface=[[sg.Text("Parâmetros do exame: idnovo, data, nome, idade, género, morada, modalidade, clube, email, federado, resultado")], 
           [sg.Button("Listar Exames"), sg.Button("Consultar Exame"),sg.Button("Lista de Modalidades"),],
           [sg.Button("Distribuições"),   sg.Button("Gráficos")],
           [sg.Text("----------------------------------------------------------------------------------------------------------------------------------------------------")],
           [sg.Listbox(values=res,size=(100,10), key="_Resultado")],
           [sg.Button("Sair")]
          ]

window = sg.Window("Exames Médicos", interface)

stop = False
while not stop:
    event, values = window.read()
    if event == "Sair" or event == sg.WIN_CLOSED:
        stop = True
        
    elif event=="Listar Exames":   
        res=[]
        window["_Resultado"].update(values=res)
        res=tpc.listarDataset(BD)
        window["_Resultado"].update(values=res)
        
    elif event=="Consultar Exame":
        res=[]
        window["_Resultado"].update(values=res)
        interid = [[sg.Text('Id do paciente: ', size=(10, 1)), sg.InputText(key='_Id')],[sg.Button('Inserir')]]

        wid = sg.Window('Inserir id', interid)
        inputEventid, inputValuesid = wid.read(close=True)
        wid.close()
        if inputEventid == 'Inserir':
            res=[]
            res.append(tpc.consultarDataset(BD,str(inputValuesid['_Id'])))
            res.append(" ")
        window["_Resultado"].update(values=res)
        
    elif event=="Lista de Modalidades":
        res=[]
        window["_Resultado"].update(values=res)
        res=tpc.modalidades(BD)
        window["_Resultado"].update(values=res)
    
    elif event=="Distribuições":
        res=[]
        window["_Resultado"].update(values=res)
        formLayout = [[sg.Button("Distribuição por modalidade")],
                 [sg.Button("Distribuição por clube")],
                 [sg.Button("Distribuição por ano")],
                 [sg.Button("Distribuição por estado de federado")],
                 [sg.Button("Distribuição por resultado do teste")],
                 [sg.Button("Distribuição por idade")],
                 [sg.Button("Distribuição por género")]]

        wform = sg.Window('Distribuições', formLayout)
        inputEvent, inputValues = wform.read(close=True)
     
        if inputEvent == "Inserir" or event == sg.WIN_CLOSED:
            stop = True 
            
        elif inputEvent=="Distribuição por ano":
            anos=tpc.distribPorAno(BD)
            res.append("Ano: Nº de exames realizados")
            for a in anos:
                res.append(str(a)+": "+str(anos[a]))
                
        elif inputEvent=="Distribuição por modalidade":
            res=(distribui(BD,6))
            res.insert(0,"Modalidade: Nº de atletas")
            
        elif inputEvent=="Distribuição por clube":
            res=distribui(BD,7)
            res.insert(0,"Clube: Nº de atletas")
                    
        elif inputEvent=="Distribuição por estado de federado":
            res=distribui(BD,9)
            res.insert(0,"Federado/Não federado: Nº de atletas")
                    
        elif inputEvent=="Distribuição por resultado do teste":
            res=distribui(BD,10)
            res.insert(0,"Apto/ Não apto: Nº de atletas")
  
        elif inputEvent=="Distribuição por idade":
            res=(distribui(BD,3))
            res.insert(0,"Idade: Nº de atletas")
            
        elif inputEvent=="Distribuição por género":  
            res=distribui(BD,4)
            res.insert(0,"Género: Nº de atletas")
            
        window["_Resultado"].update(values=res)
        
    elif event=="Gráficos":
        res=[]
        window["_Resultado"].update(values=res)
        grafLayout = [[sg.Button("Gráfico da distribuição por modalidade")],
                 [sg.Button("Gráfico da distribuição por clube")],
                 [sg.Button("Gráfico da distribuição por ano")],
                 [sg.Button("Gráfico da distribuição por estado de federado")],
                 [sg.Button("Gráfico da distribuição por resultado do teste")],
                 [sg.Button("Gráfico da distribuição por idade")],
                 [sg.Button("Gráfico da distribuição por género")]]

        wgraf = sg.Window('Distribuições', grafLayout)
        Eventgraf, Valuesgraf = wgraf.read(close=True)
 
        if Eventgraf=="Gráfico da distribuição por modalidade":
            tpc.plotDist(BD,6,"Modalidades")
        
        elif  Eventgraf=="Gráfico da distribuição por clube":
             tpc.plotDist(BD,7,"Clubes")
        
        elif Eventgraf=="Gráfico da distribuição por estado de federado":
            tpc.plotDist(BD,9,"Estado de federado")
        
        elif Eventgraf=="Gráfico da distribuição por resultado do teste":
            tpc.plotDist(BD,10,"Resultado do teste")
        
        elif Eventgraf=="Gráfico da distribuição por idade":
            tpc.plotDist(BD,3,"Idade")
            
        elif Eventgraf=="Gráfico da distribuição por género":
            tpc.plotDist(BD,4,"Género")   
            
        elif Eventgraf=="Gráfico da distribuição por ano":
            tpc.plotDistribPorAno(BD)
        
window.close()
```
