```python
def criarFrac():
    n=int(input("Insira o numerador da fração: "))
    d=int(input("Insira o denominador da fração: "))
    return (n,d)

def verFrac(f):
    n,d=f
    if d==1:
        s=str(n)
    else:
        s=str(n)+"/"+str(d)
    return s
    
def mdc(f):
    n,d=f
    while d!=0:
        resto=n%d
        n=d
        d=resto
    return n

def simplificar(f):
    n,d=f
    divisor=mdc(f)  
    simplificada=(int(n/divisor),int(d/divisor)) 
    return simplificada

def somarduas(f1,f2):
    f1nova=simplificar(f1)
    f2nova=simplificar(f2)
    n1,d1=f1nova
    n2,d2=f2nova
    f3=((n1*d2+n2*d1),(d1*d2))
    return simplificar(f3)

def somarvarias(num):
    i=1
    print("1ª Fração ")
    res=criarFrac()
    i=2
    while i<=num:
        print(str(i)+("ª Fração "))
        f1=criarFrac()
        res=somarduas(res, f1)
        i=i+1
    return res

def subtrairduas(f1,f2):
    f1nova=simplificar(f1)
    f2nova=simplificar(f2)
    n1,d1=f1nova
    n2,d2=f2nova
    f3=((n1*d2-n2*d1),(d1*d2))
    return simplificar(f3)

def subtrairvarias(num):
    i=1
    print("1ª Fração ")
    res=criarFrac()
    i=2
    while i<=num:
        print(str(i)+("ª Fração "))
        f1=criarFrac()
        res=subtrairduas(res, f1)
        i=i+1
    return res

def multiplicarduas(f1,f2):
    f1nova=simplificar(f1)
    f2nova=simplificar(f2)
    n1,d1=f1nova
    n2,d2=f2nova
    res=((n1*n2),(d1*d2))
    return simplificar(res)

def multiplicarvarias(num):
    i=1
    print("1ª Fração ")
    res=criarFrac()
    i=2
    while i<=num:
        print(str(i)+("ª Fração "))
        f1=criarFrac()
        res=multiplicarduas(res, f1)
        i=i+1
    return res

def dividirduas(f1,f2):
    f1nova=simplificar(f1)
    f2nova=simplificar(f2)
    n1,d1=f1nova
    n2,d2=f2nova
    res=(multiplicarduas(f1nova, (d2,n2)))
    return res

def dividirvarias(num):
    i=1
    print("1ª Fração ")
    res=criarFrac()
    i=2
    while i<=num:
        print(str(i)+("ª Fração "))
        f1=criarFrac()
        res=dividirduas(res, f1)
        i=i+1
    return res

def menu():
    print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
    ope=str(input("Insira a operação que pretende realizar: "))
    while ope!="0":
        if ope=="1":
            num=int(input("Quantas frações quer somar? "))
            resultado=somarvarias(num)
            print("O resultado é: ",verFrac(resultado))
            
            print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
            ope=str(input("Insira a operação que pretende realizar: "))
        elif ope=="2":
            num=int(input("Quantas frações quer subtrair? "))
            resultado=subtrairvarias(num)
            print("O resultado é: ",verFrac(resultado))    
            print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
            ope=str(input("Insira a operação que pretende realizar: "))
        elif ope=="3":
            num=int(input("Quantas frações quer multiplicar? "))
            resultado=multiplicarvarias(num)
            print("O resultado é: ",verFrac(resultado))    
            print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
            ope=str(input("Insira a operação que pretende realizar: "))
        elif ope=="4":
            num=int(input("Quantas frações quer divir? "))
            resultado=dividirvarias(num)
            print("O resultado é: ",verFrac(resultado))    
            print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
            ope=str(input("Insira a operação que pretende realizar: "))
        elif ope=="5":
            f=criarFrac()
            print(verFrac(f),"=",verFrac(simplificar(f)))
            print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
            ope=str(input("Insira a operação que pretende realizar: "))
        elif ope=="6":
            f=criarFrac()
            print("O maior fator de simplificação da fração é: ",mdc(f))
            print("""
Operações disponiveis:
(1)Somar frações;
(2)Subtrair frações;
(3)Multiplicar frações;
(4)Dividir frações;
(5)Simplificar Frações;
(6)Calcular o maior fator de simplificação da fração;
(0)Sair""")
            ope=str(input("Insira a operação que pretende realizar: "))
        elif ope=="0":
            break
        else:
            print("Erro, insira uma operação válida.")
            ope=str(input("Insira a operação que pretende realizar: "))
menu()            
```
